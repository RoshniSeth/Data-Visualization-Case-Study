###**Data Visualization on Tips Dataset**

####**Loading the Dataset**



```python
# Printing the built-in datasets in seaborn library
print(sns.get_dataset_names())

# Loading the "Tips" dataset
tips=sns.load_dataset("tips")

print()

# Printing top 5 rows of tips dataset
print(tips.head())
```

####**Preprocessing the Dataset**

```python
print("Number Of rows and columns in dataset: ",tips.shape)
print()
print("Information about the dataset: ")
print()
print(tips.info())
print("Null values in the dataset: ")
print()
print(tips.isnull().sum())
```

**Analysis**-> From the above code, we found that the dataset does not contain any null values and the column names and there datatypes are correct.There are 244 rows and 7 columns in the dataset.

####**Description About the Dataset**

The dataset typically includes the following columns:

1. **total_bill**: The total bill amount for a meal, including tax and
   excluding tips.
2. **tip**: The amount of tip given by the customer.
3. **sex:** The gender of the person paying the bill (male or female).
4. **smoker:** Whether the customer is a smoker or not (yes or no).
5. **day:** The day of the week when the meal took place (e.g., Thursday,   Friday, etc.).
6. **time:** The time of the day when the meal took place (Lunch or Dinner).
7. **size:** The number of people in the dining party.

####**Visualizations**

* **Subplots Including Pie Charts**

```python
sex=tips['sex'].value_counts()
print(f"Value Counts Of Sex Column\n {'-'*50}\n{sex}")

print()

smoker=tips['smoker'].value_counts()
print(f"Value Counts Of Fall Column\n {'-'*50}\n{smoker}")
```

```python
font_1={'family':'serif','size':10,'color':'forestgreen','weight':'bold'}

plt.figure(figsize=(12,4))
plt.suptitle("Visitors Analysis",family='serif',fontweight='bold',color='forestgreen',fontsize=25)

plt.subplot(1,2,1)
plt.pie(sex,explode=[0,0.1],labels=sex.index,autopct='%1.1f%%',shadow=True,startangle=40,pctdistance=1.3,colors=['magenta','chartreuse'])
plt.title("Record Showing Number of Customers Gender Wise",fontdict=font_1)
plt.legend(title="Gender",labels=sex.index,loc='lower left')
plt.axis('equal')


plt.subplot(1,2,2)
plt.pie(smoker,explode=[0,0.1],labels=smoker.index,autopct='%1.1f%%',shadow=True,startangle=40,pctdistance=1.3,colors=['magenta','chartreuse'])
plt.title("Record Showing Number of Smokers in Restaurant",fontdict=font_1)
plt.legend(title="Smoker",labels=smoker.index,loc='lower right')
plt.axis('equal')


plt.tight_layout()
plt.show()
```

####**Description of the Dataset**

The description of Diamonds dataset is as follows:

1. **Carat Weight:** This attribute represents the weight of the diamond,
   typically measured in carats. Carat weight is a significant factor in determining a diamond's value, as larger diamonds are generally more valuable.

2. **Cut:** The cut of a diamond refers to how well it has been cut and
   polished from its rough form. Common cut grades include "Ideal," "Premium," "Very Good," "Good," and "Fair." A well-cut diamond reflects light effectively, enhancing its brilliance and sparkle.

3. **Color:** Diamonds are graded on a color scale ranging from D
  (colorless) to Z (light yellow or brown). Colorless diamonds are considered more valuable, as they allow more light to pass through, resulting in greater brilliance.

4. **Clarity:** Clarity measures the presence of internal or external     flaws, known as inclusions and blemishes, respectively. Clarity grades include "FL" (flawless) to "I3" (included). Diamonds with fewer inclusions and blemishes are typically more valuable.

5. **Depth:** Depth refers to the height of a diamond (measured from the
   culet to the table) divided by its average girdle diameter. It's expressed as a percentage. Proper depth proportions ensure optimal light reflection and brilliance.

6. **Table:** The table percentage represents the width of the diamond's    table facet (the flat surface on the top of the diamond) divided by the diameter of the diamond. It's expressed as a percentage.

7. **Price:** This attribute represents the price of the diamond, usually measured in USD. The price of a diamond depends on various factors, including carat weight, cut, color, and clarity.
8. **x:** The length of the diamond, measured in millimeters. This dimension typically refers to the longest dimension of the diamond, which is usually measured along the top of the diamond, from one end to the other.

9. **y:** The width of the diamond, also measured in millimeters. This dimension represents the width of the diamond at its widest point, perpendicular to the length (x-axis).

10. **z:** The depth of the diamond, again measured in millimeters. This dimension refers to the depth of the diamond from the bottom (culet) to the top (table), typically measured perpendicular to the x and y dimensions.

```python
price_distribution = diamonds.groupby(['cut', 'color'])['price'].sum().unstack()
price_distribution.reset_index(inplace=True)
price_distribution_melted = price_distribution.melt(id_vars='cut', var_name='color', value_name='total_price')

plt.figure(figsize=(8, 4))
sns.barplot(x='cut', y='total_price', hue='color', data=price_distribution_melted, palette='magma')
plt.xlabel('Cut',fontsize=10,fontweight='bold',color='darkmagenta')
plt.ylabel('Total Price',fontsize=10,fontweight='bold',color='darkmagenta')
plt.title('Price Distribution by Color and Cut',color='darkmagenta',fontweight='bold',fontsize=20)
plt.xticks(color='darkmagenta',rotation=45)
plt.yticks(color='darkmagenta')
plt.gca().spines['bottom'].set_color('lightgrey')
plt.gca().spines['top'].set_color('lightgrey')
plt.gca().spines['left'].set_color('lightgrey')
plt.gca().spines['right'].set_color('lightgrey')

plt.legend(title='Color')
plt.grid(axis='y',color='gainsboro')
plt.show()
```

**Analysis**-> The stacked bar chart of price by color and cut  illustrates the distribution of diamond prices by color and cut.
